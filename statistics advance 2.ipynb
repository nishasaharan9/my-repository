**Question 1: What is hypothesis testing in statistics?**
Answer:

Hypothesis testing can be defined as a statistical tool that is used to identify if the results of an experiment are meaningful or not. It involves setting up a null hypothesis and an alternative hypothesis. These two hypotheses will always be mutually exclusive. This means that if the null hypothesis is true then the alternative hypothesis is false and vice versa. An example of hypothesis testing is setting up a test to check if a new medicine works on a disease in a more efficient manner.

**Null Hypothesis :**

The null hypothesis is a concise mathematical statement that is used to indicate that there is no difference between two possibilities. In other words, there is no difference between certain characteristics of data. This hypothesis assumes that the outcomes of an experiment are based on chance alone. Hypothesis testing is used to conclude if the null hypothesis can be rejected or not. Suppose an experiment is conducted to check if girls are shorter than boys at the age of 5. The null hypothesis will say that they are the same height.

**Alternative Hypothesis :**

The alternative hypothesis is an alternative to the null hypothesis. It is used to show that the observations of an experiment are due to some real effect. It indicates that there is a statistical significance between two possible outcomes and can be denoted For the above-mentioned example, the alternative hypothesis would be that girls are shorter than boys at the age of 5.

**Hypothesis Testing P Value :**

In hypothesis testing, the p value is used to indicate whether the results obtained after conducting a test are statistically significant or not. It also indicates the probability of making an error in rejecting or not rejecting the null hypothesis.This value is always a number between 0 and 1. The p value is compared to an alpha level, 
α
 or significance level. The alpha level can be defined as the acceptable risk of incorrectly rejecting the null hypothesis. The alpha level is usually chosen between 1% to 5%.

**Question 2: What is the null hypothesis, and how does it differ from the alternative 
hypothesis?**
Answer:

Null Hypothesis (H0): This hypothesis posits that there is no relationship or effect between the variables being studied. It serves as the default assumption that any observed differences are due to chance. For example, in a clinical trial, the null hypothesis might state that a new drug has no effect on patients compared to a placebo. 
Alternative Hypothesis (H1 or Ha): This hypothesis proposes that there is a significant relationship or effect. It is what researchers aim to support through their data. Continuing with the clinical trial example, the alternative hypothesis would state that the new drug does have an effect on patients. 

**Key Differences**

1. Nature of Claims:
Null Hypothesis: Claims no effect or relationship (e.g., H0: μ1 = μ2).
Alternative Hypothesis: Claims there is an effect or relationship (e.g., H1: μ1 ≠ μ2). 

2. Testing Objective:
Null Hypothesis: Researchers aim to reject the null hypothesis based on evidence from their data. 
Alternative Hypothesis: Researchers seek to support the alternative hypothesis by providing sufficient evidence against the null. 

3.Statistical Significance:
Null Hypothesis: If the null hypothesis is accepted, it suggests that any observed effect is due to random chance. 
Alternative Hypothesis: If the alternative hypothesis is accepted, it indicates that the observed effect is statistically significant and not due to chance. 

4. Notation:
Null Hypothesis: Denoted as H0.
Alternative Hypothesis: Denoted as H1 or Ha. 

**Conclusion**
In hypothesis testing, the null hypothesis serves as a starting point for statistical analysis, while the alternative hypothesis represents the outcome that researchers hope to demonstrate. Understanding the distinction between these two hypotheses is crucial for conducting effective research and interpreting statistical results.




**Question 3: Explain the significance level in hypothesis testing and its role in deciding 
the outcome of a test.**
Answer :

The significance level, often denoted by α (alpha), is a threshold used in hypothesis testing to determine whether to reject the null hypothesis (H₀).

Common values: 0.05, 0.01, or 0.10

It represents the maximum probability of making a Type I error — rejecting the null hypothesis when it is actually true.

**Role in Hypothesis Testing** 

Here's how it fits into the decision-making process:

You start with a null hypothesis (H₀) and an alternative hypothesis (H₁).

You collect data and calculate a test statistic (like a z-score or t-score).

You then compute a p-value, which tells you how likely your observed data is under the assumption that H₀ is true.

Compare the p-value to α:

If p-value ≤ α → Reject H₀ (evidence supports H₁)

If p-value > α → Fail to reject H₀ (not enough evidence against H₀)

**Why It Matters**
Controls risk: Choosing α helps control the risk of making a false positive conclusion.

Context-dependent: In medical studies, a lower α (like 0.01) is often used to minimize risk. In exploratory research, a higher α (like 0.10) might be acceptable.

Influences confidence: A lower α means stricter criteria for rejecting H₀, leading to more confidence in the results.

**Example**
Suppose you're testing a new drug:

H₀: The drug has no effect.

H₁: The drug has a positive effect.

α = 0.05

If your p-value is 0.03, you reject H₀ — there's statistically significant evidence that the drug works.

**Question 4: What are Type I and Type II errors? Give examples of each.**
Answer:

In hypothesis testing, we begin with two competing hypotheses:

Null hypothesis (H₀): Assumes no effect or no difference.

Alternative hypothesis (H₁): Assumes there is an effect or a difference.

When we analyze data, we make a decision: either reject H₀ or fail to reject H₀. But because we're working with samples and probabilities, there's always a chance we make the wrong decision. That’s where these errors come in:

**Type I Error (False Positive)**

1. Definition: Occurs when we reject the null hypothesis even though it is actually true.

2. In other words: We think we found a significant result, but it’s just due to random chance.

3. Probability of Type I Error: Denoted by α (alpha), which is the significance level (commonly set at 0.05 ).

**Example:**
Imagine a clinical trial testing a new drug:

H₀: The drug has no effect.

H₁: The drug improves patient health.

If the researchers reject H₀ and claim the drug works, but in reality, it doesn’t—this is a Type I error. They’ve made a false discovery.

**Type II Error (False Negative)**

1. Definition: Occurs when we fail to reject the null hypothesis even though it is actually false.

2. In other words: We miss a real effect or difference.

3. Probability of Type II Error: Denoted by β (beta).

Power of the test: Defined as 1 − β, which is the probability of correctly rejecting H₀ when it is false.

**Example:**

Same drug trial:

H₀: The drug has no effect.

H₁: The drug improves patient health.

If the researchers fail to reject H₀ and conclude the drug doesn’t work, but in reality, it does—this is a Type II error. They’ve missed a real breakthrough.

**Why These Errors Matter**

Type I errors can lead to wasted resources, misleading conclusions, or harmful decisions (e.g., approving an ineffective drug).

Type II errors can result in missed opportunities, delayed progress, or failure to act when needed (e.g., ignoring a life-saving treatment).

**Researchers often balance these risks by:**

Choosing an appropriate significance level (α).

Increasing sample size to reduce β and improve power.

Using replication and robust study designs.

**Question 5: What is the difference between a Z-test and a T-test? Explain when to use 
each.**
Answer :

**What Is a Z-Test?**
A Z-test is a statistical method used to determine whether there is a significant difference between sample and population means, or between two sample means, under the assumption that the population variance is known.

The Z-test relies on the standard normal distribution. Because this distribution is fixed and well understood, the Z-test is considered more precise—but only when certain conditions are met.

You should use a Z-test when:

The sample size is large, typically 30 or more.

The population standard deviation is known.

The data is approximately normally distributed, or the sample size is large enough for the Central Limit Theorem to apply.

For example, if you're comparing the average height of 100 adult males in a city to the national average, and you know the population standard deviation, a Z-test would be appropriate.

**What Is a T-Test?**  
A T-test is used when the population standard deviation is unknown, which is often the case in real-world scenarios. It’s especially useful for small sample sizes, where the normal approximation may not hold.

Instead of the standard normal distribution, the T-test uses the Student’s t-distribution, which is similar in shape but has heavier tails. This accounts for the extra uncertainty introduced by estimating the standard deviation from the sample itself.

You should use a T-test when:

The sample size is small, typically less than 30.

The population standard deviation is unknown.

The data is approximately normally distributed.

For instance, if you're testing whether a new teaching method improves student scores and you only have data from 15 students, and you don’t know the population standard deviation, a T-test is the better choice.

**Summary of When to Use Each**

The key difference lies in whether you know the population standard deviation and how large your sample is. Z-tests are ideal for large samples with known variance, while T-tests are designed for smaller samples or when the variance must be estimated from the data.

Both tests help you make decisions about hypotheses, but choosing the right one ensures your conclusions are statistically sound and reliable.

```python
# Question 6. Write a Python program to generate a binomial distribution with n=10 and 
# p=0.5, then plot its histogram.
import random
import matplotlib.pyplot as plt

# Parameters
n = 10       # number of trials
p = 0.5      # probability of success
size = 1000  # number of samples

# Generate binomial samples manually using random module
samples = []
for _ in range(size):
    success_count = sum(1 for _ in range(n) if random.random() < p)
    samples.append(success_count)

# Plot histogram
plt.figure(figsize=(8, 6))
plt.hist(samples, bins=range(n+2), edgecolor='black', color='lightblue', align='left')
plt.title('Binomial Distribution Histogram (n=10, p=0.5)', fontsize=14)
plt.xlabel('Number of Successes', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(range(n+1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
'''
#Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in 
Python. Show the Python code and interpret the results. 
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 
                        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 
                        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 
                        50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
'''

import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Assumed population parameters
population_mean = 50
population_std = 0.5  # known standard deviation

# Sample statistics
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# Z-statistic formula
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Output results
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Z-Score: {z_score:.4f}")
print(f"P-Value: {p_value:.4f}")

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Significant difference from population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference from population mean.")
```

```python
'''
Question 8: Write a Python script to simulate data from a normal distribution and 
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
'''

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# **Step 1: Simulate data from a normal distribution**
np.random.seed(42)  # For reproducibility
mean = 50  # Mean of the distribution
std_dev = 10  # Standard deviation
sample_size = 1000  # Number of samples

data = np.random.normal(mean, std_dev, sample_size)

# **Step 2: Calculate the 95% confidence interval for the mean**
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation
confidence_level = 0.95
alpha = 1 - confidence_level
z_critical = stats.norm.ppf(1 - alpha / 2)  # Z-score for 95% confidence

margin_of_error = z_critical * (sample_std / np.sqrt(sample_size))
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: {confidence_interval}")
# **Step 3: Plot the data**
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7, density=True)

# Overlay the normal distribution curve
x = np.linspace(min(data), max(data), 1000)
pdf = stats.norm.pdf(x, mean, std_dev)
plt.plot(x, pdf, color='red', linewidth=2, label='Normal Distribution Curve')
# Add labels and title
plt.title('Histogram of Simulated Data with Normal Distribution Curve')
plt.xlabel('Data Values')
plt.ylabel('Density')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
'''
Question 9: Write a Python function to calculate the Z-scores from a dataset and 
visualize the standardized data using a histogram. Explain what the Z-scores represent 
in terms of standard deviations from the mean.
'''

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

def calculate_and_visualize_zscores(data):
    """
    Calculate Z-scores for a dataset and visualize the standardized data using a histogram.
    
    Parameters:
        data (list or numpy array): The dataset to standardize.
    """
    # Calculate Z-scores
    z_scores = zscore(data)
    
    # Plot histogram of Z-scores
    plt.figure(figsize=(8, 5))
    plt.hist(z_scores, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
    plt.title('Histogram of Z-scores', fontsize=14)
    plt.xlabel('Z-score', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.axvline(0, color='red', linestyle='dashed', linewidth=1, label='Mean (Z=0)')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.show()
    
    return z_scores

# Example usage
data = [10, 12, 14, 15, 18, 20, 22, 25, 30, 35]
z_scores = calculate_and_visualize_zscores(data)
print("Z-scores:", z_scores)
```
